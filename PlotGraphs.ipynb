```python
import numpy as np
import matplotlib.pyplot as plt

ListColors = []
color_codes = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink',
                        'tab:gray', 'tab:olive', 'tab:cyan', 'gold', 'navy', 'darkorange', 'lime', 'crimson']
line_styles = ['-', '--', '-.', ':']

for l in line_styles:
    for c in color_codes:
        ListColors.append((c, l))  # Storing color code and line style as a tuple
```

# Effect of Chaging Pruning Rate

## CIFAR10

```python
#CIFAR10


FileName=["PruningIterations_ResNet18_CIFAR_05PR_10Epochs_L2_0005.npz",
          "PruningIterations_ResNet18_CIFAR_025PR_10Epochs_L2_0005.npz",
          "PruningIterations_ResNet18_CIFAR_01PR_10Epochs_L2_0005.npz",
          "PruningIterations_ResNet18_CIFAR_005PR_10Epochs_L2_0005.npz",
          "PruningIterations_VGG16_CIFAR_05PR_10Epochs_L2_0005.npz",
          "PruningIterations_VGG16_CIFAR_025PR_10Epochs_L2_0005.npz",
          "PruningIterations_VGG16_CIFAR_01PR_10Epochs_L2_0005.npz"]

LegendNameNotOOD=["ResNet18: 50%Pr - Non-OOD", "ResNet18: 25%Pr - Non-OOD", "ResNet18: 10%Pr - Non-OOD", "ResNet18: 5%Pr - Non-OOD",
            "VGG16: 50%Pr - Non-OOD", "VGG16: 25%Pr - Non-OOD", "VGG16: 10%Pr - Non-OOD"]

LegendNameOOD=["ResNet18: 50%Pr - OOD", "ResNet18: 25%Pr - OOD", "ResNet18: 10%Pr - OOD", "ResNet18: 5%Pr - OOD",
            "VGG16: 50%Pr - OOD", "VGG16: 25%Pr - OOD", "VGG16: 10%Pr - OOD"]

ColorList=["red", "blue", "green", "orange", "purple", "brown", "black"]

FilePath="Networks/Progress/"



for i in range(len(FileName)):
    NpzPath=FilePath+FileName[i]
    NpzData=np.load(NpzPath,allow_pickle=True)


    PruneRatio=1-NpzData["RemainingWeights_List"]
    TestAcc=NpzData["TestAcc"]
    Avg_Corrup_Acc=NpzData["Avg_Corrup_Acc_List"]

    plt.plot(PruneRatio,TestAcc, color=ColorList[i], label=LegendNameNotOOD[i],marker ='o')
    plt.plot(PruneRatio[Avg_Corrup_Acc!=-1],Avg_Corrup_Acc[Avg_Corrup_Acc!=-1], color=ColorList[i], label=LegendNameOOD[i],marker ='o',linestyle='--')

plt.title("CIFAR10 Accuracy vs Pruning Ratio")
plt.xlabel("Pruning Ratio")
plt.ylabel("Accuracy")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., frameon=False)

plt.xlim(0,1)
```
